```python
Tplanet = [273, 373] 
Tstar = [2800, 3300, 3700, 4400, 5800]
Rstar = [1.7679e8, 2.0836e8, 2.3362e8, 4.9405e8, 6.96e8]

Dh_2800K_273K = (0.25*(Rstar[0]**2)*((Tstar[0]/Tplanet[0])**4))**0.5
Dh_2800K_373K = (0.25*(Rstar[0]**2)*((Tstar[0]/Tplanet[1])**4))**0.5
Dh_3300K_273K = (0.25*(Rstar[1]**2)*((Tstar[1]/Tplanet[0])**4))**0.5
Dh_3300K_373K = (0.25*(Rstar[1]**2)*((Tstar[1]/Tplanet[1])**4))**0.5
Dh_3700K_273K = (0.25*(Rstar[2]**2)*((Tstar[2]/Tplanet[0])**4))**0.5
Dh_3700K_373K = (0.25*(Rstar[2]**2)*((Tstar[2]/Tplanet[1])**4))**0.5
Dh_4400K_273K = (0.25*(Rstar[3]**2)*((Tstar[3]/Tplanet[0])**4))**0.5
Dh_4400K_373K = (0.25*(Rstar[3]**2)*((Tstar[3]/Tplanet[1])**4))**0.5
Dh_5800K_273K = (0.25*(Rstar[4]**2)*((Tstar[4]/Tplanet[0])**4))**0.5
Dh_5800K_373K = (0.25*(Rstar[4]**2)*((Tstar[4]/Tplanet[1])**4))**0.5

Midpoint_Dh_2800K = ("{:e}".format(Dh_2800K_373K+(0.5*(Dh_2800K_273K - Dh_2800K_373K))))
Midpoint_Dh_3300K = ("{:e}".format(Dh_3300K_373K+(0.5*(Dh_3300K_273K - Dh_3300K_373K))))
Midpoint_Dh_3700K = ("{:e}".format(Dh_3700K_373K+(0.5*(Dh_3700K_273K - Dh_3700K_373K))))
Midpoint_Dh_4400K = ("{:e}".format(Dh_4400K_373K+(0.5*(Dh_4400K_273K - Dh_4400K_373K))))
Midpoint_Dh_5800K = ("{:e}".format(Dh_5800K_373K+(0.5*(Dh_5800K_273K - Dh_5800K_373K))))

Distance_Factor_2800K = "{:e}".format((Rstar[0]/float(Midpoint_Dh_2800K))**2)
Distance_Factor_3300K = "{:e}".format((Rstar[1]/float(Midpoint_Dh_3300K))**2)
Distance_Factor_3700K = "{:e}".format((Rstar[2]/float(Midpoint_Dh_3700K))**2)
Distance_Factor_4400K = "{:e}".format((Rstar[3]/float(Midpoint_Dh_4400K))**2)
Distance_Factor_5800K = "{:e}".format((Rstar[4]/float(Midpoint_Dh_5800K))**2)

print('Distance Factor for 2800K star:', Distance_Factor_2800K)
print('Distance Factor for 3300K star:', Distance_Factor_3300K)
print('Distance Factor for 3700K star:', Distance_Factor_3700K)
print('Distance Factor for 4400K star:', Distance_Factor_4400K)
print('Distance Factor for 5800K star:', Distance_Factor_5800K)
```

```python
import matplotlib.pyplot as plt #import plt function from matplotlib library
import numpy as np 
from matplotlib.ticker import (MultipleLocator, #major/minor ticks
                               FormatStrFormatter,
                               AutoMinorLocator)

res = 1

spec_wavelength2800K = [] #creating an empty array
spec_wavelength3300K = []
spec_wavelength3700K = []
spec_wavelength4400K = []
spec_wavelength5800K = []

spec_fp2800K = [] 
spec_fp3300K = []
spec_fp3700K = []
spec_fp4400K = []
spec_fp5800K = []

counter = 0 

fin2800K = open('I_2800K.txt', 'r') 
fin3300K = open('I_3300K.txt', 'r')
fin3700K = open('I_3700K.txt', 'r')
fin4400K = open('I_4400K.txt', 'r')
fin5800K = open('I_5800K.txt', 'r')

fout2800K = open('fp_2800K.txt','w') #opening the file elsewhere in directory so that you don't have to run code each time
fout3300K = open('fp_3300K.txt','w')
fout3700K = open('fp_3700K.txt','w')
fout4400K = open('fp_4400K.txt','w')
fout5800K = open('fp_5800K.txt','w')


for line in fin2800K:  
    line = line.rstrip() 
    elements = line.split('\t') 
    wavelength2800K = float(elements[0])    
    fp2800K = float(elements[1])*float(Distance_Factor_2800K)
    counter = counter + 1 
    if counter%res == 0:  
        spec_wavelength2800K.append(wavelength2800K) 
        spec_fp2800K.append(fp2800K) 
        fout2800K.write(str(wavelength2800K)+'\t'+str(fp2800K)+'\n')

        
for line in fin3300K:  
    line = line.rstrip() 
    elements = line.split('\t') 
    wavelength3300K = float(elements[0])    
    fp3300K = float(elements[1])*float(Distance_Factor_3300K)
    counter = counter + 1 
    if counter%res == 0:  
        spec_wavelength3300K.append(wavelength3300K) 
        spec_fp3300K.append(fp3300K) 
        fout3300K.write(str(wavelength3300K)+'\t'+str(fp3300K)+'\n')
        

for line in fin3700K:  
    line = line.rstrip() 
    elements = line.split('\t') 
    wavelength3700K = float(elements[0])    
    fp3700K = float(elements[1])*float(Distance_Factor_3700K)
    counter = counter + 1 
    if counter%res == 0:  
        spec_wavelength3700K.append(wavelength3700K) 
        spec_fp3700K.append(fp3700K) 
        fout3700K.write(str(wavelength3700K)+'\t'+str(fp3700K)+'\n')
        

for line in fin4400K:  
    line = line.rstrip() 
    elements = line.split('\t') 
    wavelength4400K = float(elements[0])    
    fp4400K = float(elements[1])*float(Distance_Factor_4400K)
    counter = counter + 1 
    if counter%res == 0:  
        spec_wavelength4400K.append(wavelength4400K) 
        spec_fp4400K.append(fp4400K) 
        fout4400K.write(str(wavelength4400K)+'\t'+str(fp4400K)+'\n')
        

for line in fin5800K:  
    line = line.rstrip() 
    elements = line.split('\t') 
    wavelength5800K = float(elements[0])    
    fp5800K = float(elements[1])*float(Distance_Factor_5800K)
    counter = counter + 1 
    if counter%res == 0:  
        spec_wavelength5800K.append(wavelength5800K) 
        spec_fp5800K.append(fp5800K) 
        fout5800K.write(str(wavelength5800K)+'\t'+str(fp5800K)+'\n')

    
#Plotting incidence spectra

fig, ax = plt.subplots(figsize=(15, 10)) 

#ax.set_title('Phoenix Stellar Incidence Spectra', fontsize=18) 
#ax.set_ylabel('Flux at Planet ($Wm^{-2}nm^{-1}$)', fontsize=14)
#ax.set_xlabel('Wavelength ($nm$)', fontsize=14) 

ax.set_xlim(2.0e+02, 1.25e+03)
ax.set_ylim(0.0,4.0)

ax.yaxis.set_major_locator(MultipleLocator(0.5))
ax.yaxis.set_major_formatter(FormatStrFormatter('% .1f'))
ax.tick_params(axis='both', which='major', labelsize=18)
ax.tick_params(axis='both', which='minor', labelsize=18)

ax.plot(spec_wavelength5800K,spec_fp5800K,linestyle='-',color='b',label='5800K')
ax.plot(spec_wavelength4400K,spec_fp4400K,linestyle='-',color='deepskyblue',label='4400K')
ax.plot(spec_wavelength3700K,spec_fp3700K,linestyle='-',color='g',label='3700K')
ax.plot(spec_wavelength3300K,spec_fp3300K,linestyle='-',color='tab:orange',label='3300K')
ax.plot(spec_wavelength2800K,spec_fp2800K,linestyle='-',color='r',label='2800K')

plt.fill_between(spec_wavelength5800K,spec_fp5800K,spec_fp4400K,facecolor='b',alpha=0.05)
plt.fill_between(spec_wavelength4400K,spec_fp4400K,spec_fp3700K,facecolor='deepskyblue',alpha=0.07)
plt.fill_between(spec_wavelength3700K,spec_fp3700K,spec_fp3300K,facecolor='g',alpha=0.09)
plt.fill_between(spec_wavelength3300K,spec_fp3300K,spec_fp2800K,facecolor='tab:orange',alpha=0.11)
plt.fill_between(spec_wavelength2800K,spec_fp2800K,0,facecolor='r',alpha=0.13)

plt.legend(loc=1,prop={'size':18})
plt.savefig('fp_ALLK.pdf')
plt.savefig('fp_ALLK.png') ###KEY!!! ALWAYS PUT THIS BEFORE plt.show() or you get a blank pdf.
plt.show()

#fout.close() #close fout file so it doesn't run constantly
fin2800K.close() #close fin file so it doesn't run constantly
fin3300K.close()
fin3700K.close()
fin4400K.close()
fin5800K.close()
```
