# Y1 2800K

```python
import matplotlib.pyplot as plt #import plt function from matplotlib library
import numpy as np 
from numpy import trapz
from matplotlib.ticker import (MultipleLocator, #major/minor ticks
           FormatStrFormatter, AutoMinorLocator)
import math
import pandas as pd 

hc = 1.98644568E-25

res = 1

s_2800K_1 = []
s_2800K_2 = [] 


counter = 1

with open('Y1 and Y2 photon absorption rates/2800K /fp_2800K.txt') as fp2800K, open('Y1 and Y2 photon absorption rates/2800K /g1_LHCII_spectra_interp.txt') as g1_LHCII_spectra_interp, open('Y1 and Y2 photon absorption rates/2800K /y1_2800_LHCII.txt', mode = 'w') as y1_2800_LHCII:
    for line1, line2 in zip(fp2800K, g1_LHCII_spectra_interp): #zip() aggregates your separate files in the brackets
        line1, line2 = line1.rstrip(), line2.rstrip() 
        col1, col2 = line1.split('\t'), line2.split('\t') #creates column 1 (wavelength, fp spectra) and column 2 (wavelength, LHCII absorption spectra)
        
        wavelength_2800K = float(col1[0]) #1st element in column 1 (could also do col2[0]) 
        product_2800K = float(col1[1])*float(col2[1])*((float(col1[0])*1E-9)/hc) #2nd element in column 1 (fp spectra) x 2nd element in column 2 (absorption spectra) x wavelength in SI units/hc
        
        counter = counter + 1 
                                        
        if counter%res == 0: 
            s_2800K_1.append(wavelength_2800K) 
            s_2800K_2.append(product_2800K) 
            
#plotting
fig, ax = plt.subplots(figsize=(15, 10))
ax.set_title('Photon Input Rate for g1 in LHCII Antenna around 2800K star', fontsize=20) 
ax.set_ylabel('Amplitude ($au$)', fontsize=14)
ax.set_xlabel('Wavelength ($nm$)', fontsize=14) 

ax.plot(s_2800K_1,s_2800K_2,linestyle='-',color='r')

plt.savefig('Y1 and Y2 photon absorption rates/2800K /Y1_2800K.png') 
plt.show() 

#Calculate Y1 (photon input rate for g1 domain)
int1 = np.trapz(s_2800K_2, s_2800K_1)

A1 = 0.43
A2 = 0.75
sigma = 1.498e-17 #metres squared
B = A1/A2 

y1 = sigma*B*int1
print('Photon Input Rate for LHCII g1', y1)
```
