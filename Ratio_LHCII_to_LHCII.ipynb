## Ratio Power In LHCII/LHCII

```python
import matplotlib.pyplot as plt #import plt function from matplotlib library
import numpy as np 
from numpy import trapz
from matplotlib.ticker import (MultipleLocator, #major/minor ticks
           FormatStrFormatter, AutoMinorLocator)
import math
import pandas as pd 

Pin_LHCII_2800K = 0.18503433271722686
Pin_LHCII_3300K = 0.7189461378088072
Pin_LHCII_3700K = 1.1863150236106934
Pin_LHCII_4400K = 1.9784128414225242
Pin_LHCII_5800K = 2.343148409663928

Pin_ratio_2800K = Pin_LHCII_2800K/Pin_LHCII_5800K 
Pin_ratio_3300K = Pin_LHCII_3300K/Pin_LHCII_5800K 
Pin_ratio_3700K = Pin_LHCII_3700K/Pin_LHCII_5800K 
Pin_ratio_4400K = Pin_LHCII_4400K/Pin_LHCII_5800K 
Pin_ratio_5800K = Pin_LHCII_5800K/Pin_LHCII_5800K 

print('2800K Relative Power:', Pin_ratio_2800K)
print('3300K Relative Power:', Pin_ratio_3300K)
print('3700K Relative Power:', Pin_ratio_3700K)
print('4400K Relative Power:', Pin_ratio_4400K)
print('5800K Relative Power:', Pin_ratio_5800K)

temp = [2800,3300,3700,4400,5800]
Pin_ratio = [float(Pin_ratio_2800K), float(Pin_ratio_3300K), float(Pin_ratio_3700K), float(Pin_ratio_4400K), float(Pin_ratio_5800K)]

          
#plotting
from scipy.interpolate import interp1d #1 dimensional interpolation
y_f = interp1d(temp,Pin_ratio,'cubic')
x = np.linspace(2800,5800,100) #start,end,number of interpolated points
y = y_f(x)

fig, ax = plt.subplots(figsize=(15, 10))

color = ['red','tab:orange','g','deepskyblue','b']
s = [900,900,100,100,100]
plt.scatter(temp,Pin_ratio,marker='o',c=color,s=s)
ax.plot(x,y,color='k',linestyle='solid',linewidth=2)

plt.axhline(y=1.0, color='b', linestyle='dotted',linewidth=2) 

ax.set_xlim(2700, 6000) 
ax.set_ylim(0.00, 1.20)
ax.yaxis.set_major_locator(MultipleLocator(0.2))
ax.yaxis.set_major_formatter(FormatStrFormatter('%0.1f'))
ax.xaxis.set_major_locator(MultipleLocator(400))
ax.xaxis.set_major_formatter(FormatStrFormatter('%1.0f'))
ax.tick_params(axis='both', which='major', labelsize=18)
ax.tick_params(axis='both', which='minor', labelsize=18)

plt.savefig('Ratio_LHCII_to_LHCII_ALLK.pdf') 
plt.savefig('Ratio_LHCII_to_LHCII_ALLK.png') 
plt.show() 
```
